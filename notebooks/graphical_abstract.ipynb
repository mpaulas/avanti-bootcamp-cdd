```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/atlantico-academy/datasets/refs/heads/main/diamonds.csv"
diamonds = pd.read_csv(url)
diamonds_100 = diamonds.head(100)

print(diamonds_100)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=diamonds_100, x='carat', y='price', hue='cut', alpha=0.5)

plt.title('Relação entre Carat e Preço dos 100 primeiros Diamantes')
plt.xlabel('Peso (Carat)')
plt.ylabel('Preço (USD)')

plt.savefig('graphical_diamonds.png')

plt.legend(title='Corte')
plt.grid()
plt.show()
```
